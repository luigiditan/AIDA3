```python
import numpy as np 

x = np.array([1, 0])
w1 = np.array([[1, -1],
               [-1, 1]]) 
w2 = np.array([[1, 2],
               [2, 1]])
w3 = np.array([-1, 1])
```

```python
z1 = x @ w1
z1
```

```python
a1 = np.tanh(z1)
a1
```

```python
z2 = a1 @ w2
z2
```

```python
a2 = np.tanh(z2)
a2
```

```python
z3 = a2 @ w3
z3
```

```python
a3 = np.tanh(z3)
a3
```

```python
y = np.tanh(a3)
y
```

```python
d_n = 2.86
o_n = y
error_signal = d_n - o_n
error_signal
```

```python
lr = 0.25
dr = (1-y)*(1+y)
delta = error_signal * dr
w_1 = w3 - (lr * delta) * a3
w_1
```

```python
dr1 = (1-a3)*(1+a3)
delta1 = (w2 * dr1) * delta
delta1
```

```python
w_2 = w2 - (lr * delta) * delta1
w_2
```
